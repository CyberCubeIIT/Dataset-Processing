# Load Dataset

```python
# import Pandas and Regex Libs
import pandas as pd
import re
```

```python
# nRow = number of rows need to process
nRowsRead = 50
# Read csv file and store in Pandas DataFrame
df = pd.read_csv('xxxxx.csv', delimiter=',', nrows = nRowsRead)
df.dataframeName = 'xxxxx.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
# Print first 5 rows with headers
df.head(5)
```

```python
# Print first row with all the headers
df.iloc[0]
```

```python
# Keep only required Columns and drop others
recipe_cols = ['RecipeId', 'Name', 'AuthorName', 'CookTime', 'PrepTime', 'TotalTime', 'Description', 'Images', 'RecipeCategory', 'Keywords', 'RecipeIngredientQuantities', 'RecipeIngredientParts', 'AggregatedRating', 'Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent', 'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent', 'RecipeServings', 'RecipeInstructions']
rp_df = df[recipe_cols]
```

```python
# Drop all the Dupliacte rows
rp_df = rp_df.drop_duplicates('RecipeId')
```

```python
# Print dataFrame info
rp_df.info()
```

```python
# Drop all the rows with Image column as "character(0)"
rp_df = rp_df[rp_df.Images != 'character(0)']
rp_df.info()
```

```python
# Functions to convert R vectors to arrays
def map_str_to_list(string):
    #pattern = re.compile(r'\"(.+)\"')
    pattern = re.compile(r'\"([^"]+)\"')
    return pattern.findall(string)

def map_for_series(series: pd.Series):
    return series.apply(lambda i: map_str_to_list(i))
```

```python
# Convert to arrays only selected columns
modify_col_list = ['Images', 'Keywords', 'RecipeIngredientQuantities', 'RecipeIngredientParts', 'RecipeInstructions']
rp_df = rp_df.apply(lambda x: map_for_series(x) if x.name in modify_col_list else x)
```

```python
# save dataFrame as csv
rp_df.reset_index().to_csv('recipes_50_processed.csv', index = False)
```
