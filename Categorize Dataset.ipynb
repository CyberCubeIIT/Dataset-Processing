# Categorize Recipes into food type

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('recipes_processed.csv')
```

```python
df.head()
```

```python
df.drop('index', axis=1, inplace=True)
df.info()
```

```python
df['FoodTypes'] = np.nan
df['FoodTypes'] = df['FoodTypes'].astype('str')
```

```python
for i in df['RecipeIngredientParts'].index:
    if('eggs' not in df['RecipeIngredientParts'][i]):
         if('ice-cream' in df['RecipeIngredientParts'][i] or 'chocolate' in df['RecipeIngredientParts'][i] or 'cookies' in df['RecipeIngredientParts'][i]):
                df['FoodTypes'][i]='Veg dessert'
    elif('eggs' in df['RecipeIngredientParts'][i]):
        if('ice-cream' in df['RecipeIngredientParts'][i] or 'chocolate' in df['RecipeIngredientParts'][i] or 'cookies' in df['RecipeIngredientParts'][i]):
                df['FoodTypes'][i]='Non-Veg dessert'
```

```python
for i in df.index:
    if(df['FoodTypes'][i]!='Veg dessert' and df['FoodTypes'][i]!='Non-Veg dessert' and 20<df['Calories'][i]<300):
        df['FoodTypes'][i]='Healthy'
```

```python
for i in df.index:
    if(df['FoodTypes'][i]!='Veg dessert' and df['FoodTypes'][i]!='Non-Veg dessert' and df['FoodTypes'][i]!='Healthy'):
        if('chicken' in df['RecipeIngredientParts'][i] or 'eggs' in df['RecipeIngredientParts'][i] or'ham' in df['RecipeIngredientParts'][i] or 'pepperoni' in df['RecipeIngredientParts'][i] ):
            df['FoodTypes'][i]='Non-veg'
```

```python
for i in df.index:
    if(df['FoodTypes'][i]!='Veg dessert' and df['FoodTypes'][i]!='Non-Veg dessert' and df['FoodTypes'][i]!='Healthy' and df['FoodTypes'][i]!='Non-veg'):
        df['FoodTypes'][i]='Veg'
```

```python
types = pd.get_dummies(df['FoodTypes'])
```

```python
df1 = pd.concat([df,types],axis = 1)
```

```python
df1.head()
```

```python
df1['Index'] = range(1, len(df1) + 1)
df1.insert(0, 'Index', df1.pop('Index'))
df1.to_csv('recipes_categorized.csv', index = False)
```
